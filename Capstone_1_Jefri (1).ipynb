# Processing Data

## Requirements Data

```python
# Null checking
requirements_data.isnull().sum()
```

```python
# Drop column job_title and job_description
requirements_data.drop(columns=['job_title', 'job_description'], inplace=True)
```

```python
# Fungsi untuk menyelaraskan data
def convert_salary_estimate(salary_str):
    # Handle cases where salary_str is not a string (e.g., already a float)
    if not isinstance(salary_str, str):
        return salary_str  # Return as is if it's already a number

    # Menghilangkan simbol dan mengubah ke format standar
    salary_str = salary_str.replace('$', '').replace('(est.)', '').strip()

    if '/yr' in salary_str:
        # Jika tahun, cukup bersihkan dan konversi ke angka
        return float(salary_str.replace('/yr', '').replace(',', '').strip())
    elif '/hr' in salary_str:
        # Jika per jam, konversi ke tahunan
        hourly_rate = float(salary_str.replace('/hr', '').replace(',', '').strip())
        annual_salary = hourly_rate * 40 * 52  # Menghitung gaji tahunan
        return annual_salary
    else:
        return None  # Jika format tak dikenal
```

```python
# Renaming columns
requirements_data.rename(columns={
    'salary_estimate': 'salary_estimate($/yr)',
}, inplace=True)
```

```python
requirements_data['salary_estimate($/yr)'] = requirements_data['salary_estimate($/yr)'].apply(convert_salary_estimate)
```

```python
requirements_data
```

```python
requirements_data['salary_estimate($/yr)'].fillna(0, inplace=True)
```

```python
requirements_data['company_founded'].fillna(0, inplace=True)
```

```python
requirements_data['company_rating'].fillna(0, inplace=True)
```

```python
# Fill NaN values in specified columns with "Unknown / Non-Applicable"
columns_to_fill = ['company_size', 'company_type', 'company_sector', 'company_industry', 'company_founded', 'company_revenue']
requirements_data[columns_to_fill] = requirements_data[columns_to_fill].fillna("Unknown / Non-Applicable")
```

```python
# Change format to GMT +7 (Asia/Jakarta)

# Ensure 'dates' column is of datetime type before localizing
requirements_data['dates'] = pd.to_datetime(requirements_data['dates'], errors='coerce')


# Localize only if the value is a valid datetime object
requirements_data['dates'] = requirements_data['dates'].apply(
    lambda x: x.tz_convert('Asia/Jakarta') if pd.notna(x) else x  # Changed 'infer' to 'NaT'
)

requirements_data['dates']
```

```python
# Delete the first column of requirements_data
requirements_data = requirements_data.iloc[:, 1:]
```

```python
requirements_data.info()
```

## Air Conditioners Data

```python
# Null checking
air_conditioners.isnull().sum()
```

```python
# Round ratings to 1 decimal place
air_conditioners['ratings'] = air_conditioners['ratings'].round(1)

air_conditioners
```

```python
# Convert 'ratings' column to float
air_conditioners['ratings'] = pd.to_numeric(air_conditioners['ratings'], errors='coerce')
```

```python
# Convert 'no_of_ratings' column to float
air_conditioners['no_of_ratings'] = pd.to_numeric(air_conditioners['no_of_ratings'], errors='coerce')
```

```python
# Replace NaN values with 0 in specified columns
air_conditioners['ratings'].fillna(0, inplace=True)
air_conditioners['no_of_ratings'].fillna(0, inplace=True)
air_conditioners['discount_price(₹)'].fillna(0, inplace=True)
air_conditioners['actual_price(₹)'].fillna(0, inplace=True)
```

```python
# Renaming columns
air_conditioners.rename(columns={
    'discount_price': 'discount_price(₹)',
    'actual_price': 'actual_price(₹)'
}, inplace=True)
```

```python
# Removing ₹ and converting to integer
air_conditioners['discount_price(₹)'] = air_conditioners['discount_price(₹)'].replace({'₹': '', ',': ''}, regex=True).astype(float)
air_conditioners['actual_price(₹)'] = air_conditioners['actual_price(₹)'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```
